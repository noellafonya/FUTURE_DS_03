```python
df['average_percent'] = df['average_percent'].astype(str).str.split('/').str[0]
df['average_percent'] =pd.to_numeric(df['average_percent'],errors='coerce')
df['average_percent'].head(50)
```

```python
course_scores = df.groupby('course_name')['average_percent'].mean().reset_index()
course_scores = course_scores.sort_values(by='average_percent', ascending=False)

plt.figure(figsize=(10,6))
sns.barplot(x='average_percent', y='course_name', data=course_scores, palette='crest')
plt.title("Average Student Satisfaction by Course", fontsize=14)
plt.xlabel("Average Percentage")
plt.ylabel("Course Name")
plt.show()
```
